```python
# _*_ coding: UTF-8 _*_
from urllib.request import urlopen
from bs4 import BeautifulSoup
import csv
```

```python
html = urlopen('https://www.imdb.com/chart/top')
bsObject = BeautifulSoup(html, "html.parser")
bsObject
```

```python
for n in range(0,2):
    td = bsObject.find_all("td", {"class":"titleColumn"})[n].get_text()
    print(td)
```

```python
for n in range(0,50):
    td = bsObject.find_all("td", {"class":"titleColumn"})[n].get_text()
    b = bsObject.find_all("td", {"class":"ratingColumn imdbRating"})[n].get_text()
    a=td.split("  ") 
    title=a[-1]
    rating=b
    print(n+1,title,rating)
```

```python
with open('imdb rating.csv', 'w') as f:
    writer = csv.writer(f)
    writer.writerow(['rank', 'title', 'rating'])
        
    for n in range(0,20):
        td = bsObject.find_all("td", {"class":"titleColumn"})[n].get_text()
        b = bsObject.find_all("td", {"class":"ratingColumn imdbRating"})[n].get_text()
        a=td.split("  ") 
        title=a[-1]
        rating=b
        writer.writerow([n+1,title,rating])
```
